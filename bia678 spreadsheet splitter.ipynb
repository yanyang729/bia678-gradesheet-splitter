```python
import pandas as pd
```

```python
df = pd.read_excel('./complete grading spreadsheet distribution v1.xlsx',header=1,sheetname='BIA678B')
```

```python
df
```

```python
names = df.Name
letter_grades = df['Letter Grade ']
oral_comms = df['Oral Report Comments']
report_comms = df['Report Comments']
overall_comms = df['Overall Comments']
```

```python
from collections import OrderedDict
for name,letter_grade, report_comm ,oral_comm, overall_comm in zip(names, letter_grades, report_comms,oral_comms, overall_comms):
    d = {
        'Letter Grade': [letter_grade],
        'Oral Report Comments':[oral_comm],
        'Report Comments':[report_comm],
        'Overall Comments': [overall_comm]
    }
    d= OrderedDict(d)
    df_out = pd.DataFrame(d).transpose()
    df_out.to_excel(name+' grade and comment.xlsx',header=None)
```
